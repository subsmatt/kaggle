```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
```

```python
# Add new features
# -----------------------------------------------------------------------------------------------------------------------------
# Knowing that this is a card game (poker) and the aim is to classify a hand of five cards without coding the rules explicitely
# let's add some new features:
# - five pairs "rank,suit" rather then having separate features S1,C1,etc. and name them P1,..,P5 such that rank is represented
#   by hex number
```

```python
train['P1'] = ''
train['P2'] = ''
train['P3'] = ''
train['P4'] = ''
train['P5'] = ''

train['NR1'] = ''
train['NS1'] = ''
train['NR2'] = ''
train['NS2'] = ''
train['NR3'] = ''
train['NS3'] = ''
train['NR4'] = ''
train['NS4'] = ''
train['NR5'] = ''
train['NS5'] = ''

train['ranks'] = ''
train['suits'] = ''
```

```python
for i in range(0,len(train)):
    train['P1'][i] = (train['C1'][i], train['S1'][i]) #(hex(train['C1'][i]).lstrip('0x').rstrip('L'),hex(train['S1'][i]).lstrip('0x').rstrip('L'))
    train['P2'][i] = (train['C2'][i], train['S2'][i]) #(hex(train['C2'][i]).lstrip('0x').rstrip('L'),hex(train['S2'][i]).lstrip('0x').rstrip('L'))
    train['P3'][i] = (train['C3'][i], train['S3'][i]) #(hex(train['C3'][i]).lstrip('0x').rstrip('L'),hex(train['S3'][i]).lstrip('0x').rstrip('L'))
    train['P4'][i] = (train['C4'][i], train['S4'][i]) #(hex(train['C4'][i]).lstrip('0x').rstrip('L'),hex(train['S4'][i]).lstrip('0x').rstrip('L'))
    train['P5'][i] = (train['C5'][i], train['S5'][i]) #(hex(train['C5'][i]).lstrip('0x').rstrip('L'),hex(train['S5'][i]).lstrip('0x').rstrip('L'))
train.head(5)
```

```python
train.drop('S1',axis=1, inplace=True)
train.drop('C1',axis=1, inplace=True)
train.drop('S2',axis=1, inplace=True)
train.drop('C2',axis=1, inplace=True)
train.drop('S3',axis=1, inplace=True)
train.drop('C3',axis=1, inplace=True)
train.drop('S4',axis=1, inplace=True)
train.drop('C4',axis=1, inplace=True)
train.drop('S5',axis=1, inplace=True)
train.drop('C5',axis=1, inplace=True)
train.head(5)
```

```python
for i in range(0,len(train)):
    vec = sorted([train['P1'][i], train['P2'][i], train['P3'][i], train['P4'][i], train['P5'][i]], reverse=True)
    train['P1'][i] = vec[0]
    train['P2'][i] = vec[1]
    train['P3'][i] = vec[2]
    train['P4'][i] = vec[3]
    train['P5'][i] = vec[4]
    
    train['NR1'][i] = train['P1'][i][0]
    train['NS1'][i] = train['P1'][i][1]
    train['NR2'][i] = train['P2'][i][0]
    train['NS2'][i] = train['P2'][i][1]
    train['NR3'][i] = train['P3'][i][0]
    train['NS3'][i] = train['P3'][i][1]
    train['NR4'][i] = train['P4'][i][0]
    train['NS4'][i] = train['P4'][i][1]
    train['NR5'][i] = train['P5'][i][0]
    train['NS5'][i] = train['P5'][i][1]
    
    train['ranks'][i] = ''.join([hex(train['P1'][i][0]).lstrip('0x').rstrip('L'), hex(train['P2'][i][0]).lstrip('0x').rstrip('L'), hex(train['P3'][i][0]).lstrip('0x').rstrip('L'), hex(train['P4'][i][0]).lstrip('0x').rstrip('L'), hex(train['P5'][i][0]).lstrip('0x').rstrip('L')])
    train['suits'][i] = ''.join([str(train['P1'][i][1]), str(train['P2'][i][1]), str(train['P3'][i][1]), str(train['P4'][i][1]), str(train['P5'][i][1])])
```
